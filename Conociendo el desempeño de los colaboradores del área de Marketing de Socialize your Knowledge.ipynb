```python
import streamlit as st
import pandas as pd
from PIL import Image
from vega_datasets import data as vega_data
import altair as alt


#1.El código que contenga las instrucciones para el despliegue de un título y una breve descripción de la aplicación web.
st.title("""Conociendo el desempeño de los colaboradores del área de Marketing de Socialize your Knowledge""")

#2.El código que permita desplegar el logotipo de la empresa en la aplicación web.
image =Image.open('Image_logo.png')
st.image(image)

df =pd.read_csv('employee_data.csv')
#3.El código que permita desplegar un control para seleccionar el género del empleado.
selected_gen = st.sidebar.multiselect("Select Gender", df['gender'].unique(),df['gender'].unique()) 


#4.El código que permita desplegar un control para seleccionar un rango del puntaje de desempeño del empleado.
# Add a slider to the sidebar:

_min_value =int(df['performance_score'].min())
_max_value =int(df['performance_score'].max())

selected_perfomance = st.sidebar.slider('Puntaje de desempeño',min_value=_min_value,max_value=_max_value,value=(_max_value))


#5.El código que permita desplegar un control para seleccionar el estado civil del empleado.
selected_class = st.sidebar.multiselect("estado civil", df["marital_status"].unique(),df["marital_status"].unique())

#df_result = 


df_result =df[['name_employee','birth_date','age','gender','marital_status', 
 'hiring_date' ,'position' ,'salary','performance_score','last_performance_date', 'average_work_hours' ,
 'satisfaction_level','absences']]

#6.El código que permita mostrar un gráfico en donde se visualice la distribución de los puntajes de desempeño.
df_filter =df_result.query("(gender==@selected_gen) and  (marital_status == @selected_class) and (performance_score <= @selected_perfomance)")

chart_perfromance = ( 
      alt.Chart(df_filter).mark_bar(size=50).encode(

    alt.X('performance_score'),
    alt.Y('count()')
    #alt.Size('US_Gross') 
    ).properties(width=200)
) 

st.write('Distribución de los puntajes de desempeño')
st.altair_chart(chart_perfromance, use_container_width=True)

#7. El código que permita mostrar un gráfico en donde se visualice el promedio de horas trabajadas por el género del empleado.
df_prom_horas_trabajadas = df_filter[['gender','average_work_hours']]
df_prom_horas_trabajadas = df_prom_horas_trabajadas.groupby('gender').mean().reset_index()


chart_promedio_horas = ( 
    alt.Chart(df_prom_horas_trabajadas).mark_bar(size=50).encode(

    alt.X('gender'),
    alt.Y('average_work_hours')
    ).properties(width=200)
) 

st.write('Promedio de horas trabajadas por el género del empleado')
st.altair_chart(chart_promedio_horas, use_container_width=True)


#8.El código que permita mostrar un gráfico en donde se visualice la edad de los empleados con respecto al salario de estos.

df_edad_salarios = df_filter[['age','salary']]
df_edad_salarios = df_edad_salarios.groupby('age').mean().reset_index()


chart_edad_salarios = ( 
    alt.Chart(df_edad_salarios).mark_trail().encode(

    alt.X('age'),
    alt.Y('salary')
    #alt.Size('US_Gross') 
    ).properties(width=200)
) 

st.write('Edad de los empleados con respecto al salario de estos')
st.altair_chart(chart_edad_salarios, use_container_width=True)

#9.El código que permita mostrar un gráfico en donde se visualice la relación del promedio de horas trabajadas versus el puntaje de desempeño.
df_relacion_promedio_horas_puntaje = df_filter[['performance_score','average_work_hours']]


chart_relacion_horast_vs_pdesempeno = ( 
alt.Chart(df_relacion_promedio_horas_puntaje).mark_point(size=20).encode(

   alt.X('average_work_hours'),

   alt.Y('performance_score')

) )

st.write('Relación del promedio de horas trabajadas versus el puntaje de desempeño.')
st.altair_chart(chart_relacion_horast_vs_pdesempeno , use_container_width=True)

# 10.El código que permita desplegar una conclusión sobre el análisis mostrado en la aplicación web.
st.write('''
### Conclusión
El valor que mas se obtiene en los puntajes de desempeño es de 3. 
No podemos asegurar que se tenga una relacion entre el puntaje de desempeño obtenido y el promedio de horas trabajadas.
''')
```
